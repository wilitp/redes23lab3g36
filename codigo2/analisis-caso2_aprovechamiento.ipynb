```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sb
import pandas as pd

data01 = pd.read_csv("./results/caso2_0.1.csv")

# Usar solo si la celda que queres contiene un vector, sino no tiene sentido
def get_vec01(col: str, row: int): return np.array(list(map(float, data01[col].loc[[row]].to_list()[0].split(" "))))

# Valores
enviados01 = get_vec01("vecvalue", 27)
recibidos01 = get_vec01("vecvalue", 31)
btx01 = get_vec01('vecvalue', 28)


data02 = pd.read_csv("./results/caso2_0.2.csv")

# Usar solo si la celda que queres contiene un vector, sino no tiene sentido
def get_vec02(col: str, row: int): return np.array(list(map(float, data02[col].loc[[row]].to_list()[0].split(" "))))

# Valores
enviados02 = get_vec02("vecvalue", 27)
recibidos02 = get_vec02("vecvalue", 31)
btx02 = get_vec02('vecvalue', 28)


data03 = pd.read_csv("./results/caso2_0.3.csv")

# Usar solo si la celda que queres contiene un vector, sino no tiene sentido
def get_vec03(col: str, row: int): return np.array(list(map(float, data03[col].loc[[row]].to_list()[0].split(" "))))

# Valores
enviados03 = get_vec03("vecvalue", 27)
recibidos03 = get_vec03("vecvalue", 31)
btx03 = get_vec03('vecvalue', 28)



aprovechamiento = [
    len(recibidos01)/(len(enviados01)-btx01[-1]),
    len(recibidos02)/(len(enviados02)-btx02[-1]),
    len(recibidos03)/(len(enviados03)-btx03[-1])
    ]
intervalos = [0.1, 0.2, 0.3]

plt.plot(intervalos, aprovechamiento, "ro-")
plt.title('Aprovechamiento segun intervalos de generación')
plt.xlabel("Intervalo de generación")
plt.ylabel("Aprovechamiento de carga")
plt.legend()
plt.show()
```
