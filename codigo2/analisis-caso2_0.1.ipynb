```python
# Vectores
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sb
import pandas as pd


data = pd.read_csv("./results/caso2_0.1.csv")

# Usar solo si la celda que queres contiene un vector, sino no tiene sentido
def get_vec(col: str, row: int): return np.array(list(map(float, data[col].loc[[row]].to_list()[0].split(" "))))

# Valores
bq2 = get_vec("vecvalue", 26)
enviados = get_vec('vecvalue', 27)
btx = get_vec('vecvalue', 28)
bq1 = get_vec('vecvalue', 29)
brx  = get_vec('vecvalue', 30)
recibidos = get_vec("vecvalue", 31)
delay = get_vec("vecvalue", 32)



# Tiempos
tq2 = get_vec("vectime", 26)
tenviados = get_vec('vectime', 27)
ttx = get_vec('vectime', 28)
tq1  = get_vec('vectime', 29)
trx = get_vec('vectime', 30)
trecibidos = get_vec("vectime", 31)
tdelay = get_vec("vectime", 32)
```

## Buffers en el tiempo

```python
import matplotlib.pyplot as plt
import builtins

plt.plot(ttx ,btx, color='tab:blue', label="Buf Tx")
plt.plot(tq1, bq1, color='tab:orange', label="Buf queue1")
plt.plot(tq2, bq2, color='tab:red', label="Buf queue2")
plt.plot(trx, brx, color='tab:green', label="Buf Rx")
plt.title('Ocupacion de buffers en el sistema')
plt.xlabel("tiempo de simulacion")
plt.ylabel("Cantidad de paquetes en el buffer")
plt.legend()
plt.show()


```

## Paquetes enviados/recibidos/perdidos


```python

import matplotlib.pyplot as plt
import numpy as np
import seaborn as sb
import pandas as pd

plt.plot(tenviados, enviados, color='tab:blue', label="Paquetes enviados")
plt.plot(trecibidos, recibidos, color='tab:orange', label="Paquetes recibidos")
plt.title('Paquetes envaidos/recibidos/perdidos')
plt.xlabel("Tiempo de simulacion")
plt.ylabel("Cantidad de paquetes")
plt.legend()
plt.show()


```

## Delay

```python
plt.plot(tdelay, delay, color='tab:blue', label="Segundos de delay")
plt.title('Delay')
plt.xlabel("Tiempo de simulacion")
plt.ylabel("Segundos")
plt.legend()
plt.show()
```
