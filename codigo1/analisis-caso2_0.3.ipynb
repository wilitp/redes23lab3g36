```python
# Vectores
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sb
import pandas as pd

sb.set_context(context='talk', font_scale=0.7)
data = pd.read_csv("./results/caso2_0.3.csv")

# Usar solo si la celda que queres contiene un vector, sino no tiene sentido
def get_vec(col: str, row: int): return np.array(list(map(float, data[col].loc[[row]].to_list()[0].split(" "))))

# Valores
enviados = get_vec("vecvalue", 26)
btx = get_vec('vecvalue', 27)
bq = get_vec('vecvalue', 28)
brx = get_vec('vecvalue', 29)
recibidos = get_vec("vecvalue", 30)
delay = get_vec("vecvalue", 31)



# Tiempos
tenviados = get_vec("vectime", 26)
ttx = get_vec('vectime', 27)
tq = get_vec('vectime', 28)
trx = get_vec('vectime', 29)
trecibidos = get_vec("vectime", 30)
tdelay = get_vec("vectime", 31)
```

## Buffers en el tiempo

```python
import matplotlib.pyplot as plt

plt.plot(ttx ,btx, color='tab:blue', label="Buf Tx")
plt.plot(tq, bq, color='tab:orange', label="Buf queue")
plt.plot(trx, brx, color='tab:green', label="Buf Rx")
plt.title('Ocupacion de buffers en el sistema')
plt.xlabel("tiempo de simulacion")
plt.ylabel("Cantidad de paquetes en el buffer")
plt.legend()
plt.show()


```

## Paquetes enviados/recibidos/perdidos


```python

import matplotlib.pyplot as plt

plt.plot(tenviados, enviados, color='tab:blue', label="Paquetes enviados")
plt.plot(trecibidos, recibidos, color='tab:orange', label="Paquetes recibidos")
plt.title('Paquetes envaidos/recibidos/perdidos')
plt.xlabel("Tiempo de simulacion")
plt.ylabel("Cantidad de paquetes")
plt.legend()
plt.show()


```

## Delay

```python
plt.plot(tdelay, delay, color='tab:blue', label="Retraso")
plt.title('Delay')
plt.xlabel("Tiempo de simulacion")
plt.ylabel("Segundos")
plt.legend()
plt.show()
```
